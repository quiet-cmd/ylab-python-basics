```python
from itertools import permutations
```

```python
def distance(point_1, point_2):
    return ((point_2[0] - point_1[0]) ** 2 + (point_2[1] - point_1[1]) ** 2) ** 0.5

def point_str(point, num=0, symbol='->'):
    return f"{point}[{num}] {symbol} "
```

```python
points_list = [(0,2),(2,5), (5,2), (6,6), (8,3)]
start = points_list[0]
size = len(points_list) + 1
permutations_list = list(map(lambda x: [start, *x, start], permutations(points_list[1:]) ))
result_list = []

print(*permutations_list, sep='\n')
```

```python
for r in list(permutations_list):
    distance_sum, temp = 0, point_str(start)
    for i in range(1, size):
        distance_sum += distance(r[i-1], r[i])
        temp += point_str(r[i], distance_sum) if i != size - 1 else (point_str(r[i], distance_sum, '=') + str(distance_sum))
    result_list.append((distance_sum, temp))

print(*result_list, sep='\n')
```
